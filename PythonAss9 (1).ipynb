Q1. What is a database? Differentiate between SQL and NoSQL databases.

Ans.
A database is an organized collection of structured information, or data, typically stored electronically in a computer system. 

1. SQL pronounced as “S-Q-L” or as “See-Quel” is primarily called RDBMS or Relational Databases, whereas NoSQL is a Non-relational or Distributed Database.

2. Comparing SQL vs NoSQL databases, SQL databases are table-based databases, whereas NoSQL databases can be document-based, key-value pairs, and graph databases.

3. SQL databases are vertically scalable, while NoSQL databases are horizontally scalable.

4. SQL databases have a predefined schema, whereas NoSQL databases use a dynamic schema for unstructured data.

5. Comparing NoSQL vs SQL performance, SQL requires specialized DB hardware for better performance while NoSQL uses commodity hardware.

Q5. Explain Primary Key and Foreign Key.

Ans.
1. Primary Key (PK):
A Primary Key is a column or a set of columns in a database table that uniquely identifies each row in the table.
It ensures that every record in the table is distinct and can be uniquely identified.

2. Foreign Key (FK):
A Foreign Key is a column or a set of columns in a table that establishes a link between the data in two related tables. It creates a referential integrity constraint, ensuring that the values in the Foreign Key column(s) of one table correspond to the values in the Primary Key column(s) of another table.
This relationship represents a "parent-child" relationship between the tables.
