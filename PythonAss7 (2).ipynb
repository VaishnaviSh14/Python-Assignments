Q1. What is an Exception in python? Write the difference between Exceptions and syntax errors Ans.

In Python, an exception is a runtime error that occurs during the execution of a program. When an exceptional situation arises, Python raises an exception, indicating that something unexpected or erroneous has happened. Exceptions allow the program to handle errors gracefully and provide a way to propagate error information up the call stack.

Differences between Exceptions and Syntax Errors:

Occurrence: Exception: Exceptions occur during the runtime of a program when an error condition is encountered while executing code. Syntax Error: Syntax errors, also known as parsing errors, occur when the Python interpreter encounters code that does not follow the proper syntax rules of the language. These errors happen before the code is executed. Handling:

Exception: Exceptions can be caught and handled using try-except blocks, allowing the program to gracefully recover from the error and continue its execution. Syntax Error: Syntax errors cannot be caught or handled within the program. They need to be fixed by the developer before the code can be successfully executed. Examples:

Exception: Examples of exceptions include ZeroDivisionError, TypeError, IndexError, ValueError, etc. For instance, trying to divide a number by zero or accessing an index that doesn't exist in a list would raise exceptions. Syntax Error: Examples of syntax errors include forgetting a colon at the end of an if statement, using an invalid character in a variable name, or mismatched parentheses and syntax errors

```python
Q3.Which Python statements are used to manage and handle exceptions? Explain with an example.
Ans.

In Python, try, except, else, and finally statements are used to manage and handle exceptions. 
These statements allow you to gracefully handle errors and implement error recovery mechanisms in your code.

Eg-
try:
    f=open("test.txt",'w')
    f.write("Hello")
except Exception as e:
    print("There is some issue" + e)
else:
    print("This block will execute once try will execute itself without an exception")
finally:
    print("This block will always execute")
    f.close()
```

```python
Q4. Explain with an example.
a. try and else
b. finally
c. raise

Ans.
a.
In Python, the try and else blocks are used for handling exceptions in code.
An exception is an error that occurs during the execution of a program, disrupting the normal flow of the program. 
Using these blocks allows you to handle potential errors gracefully and take appropriate actions.

try:
    f=open("test.txt",'w')
    f.write("Hello")
except Exception as e:
    print("There is some issue" + e)
else:
    f.close()
    
b.
The finally block is used in conjunction with try and except blocks and provides a way to ensure that some code is executed no matter what, 
whether an exception occurred or not.

try:
    f=open("test.txt",'w')
    f.write("Hello")
except Exception as e:
    print("There is some issue" + e)
else:
    print("This block will execute once try will execute itself without an exception")
finally:
    print("This block will always execute")
    f.close()
    
c.
The raise keyword in Python is used to manually raise exceptions. 
It allows you to trigger specific exceptions when certain conditions are met.

class validateage(Exception):
    def __int__(self,msg):
        self.msg = msg

def validate_age(age):
        if age<0:
            raise validateage("Age shouldn't be less that 0")
        elif age>200:
            raise validateage("Age too high")
        else:
            print("Age is valid")
                   
try:
    age = input("Enter your age")
    validate_age(age)
except validate_age as e:
    print(e)
```
