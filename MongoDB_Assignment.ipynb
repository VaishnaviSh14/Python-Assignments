Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use
MongoDB over SQL databases?

Ans.
MongoDB is a popular open-source NoSQL database management system that falls under the category of non-relational databases. It was developed to handle large volumes of unstructured or semi-structured data efficiently. MongoDB stores data in a format called BSON, which is a binary representation of JSON-like documents.

Non-relational databases, also known as NoSQL databases, differ from traditional SQL databases (relational databases) in their data storage and retrieval model. In short, non-relational databases don't use the structured tabular format with fixed schemas that SQL databases do. Instead, they offer more flexibility and scalability by allowing the storage of various data formats within the same database.

In non-relational databases like MongoDB, data is often organized into collections rather than tables, and each document can have different fields, unlike fixed columns in SQL tables. This structure is particularly useful for handling diverse and dynamic data, making it easier to adapt to changing data requirements.

Scenarios where MongoDB is preferred over SQL databases:

1. Unstructured or Semi-Structured Data

2. Scalability and Performance

3. Agile and Rapid Development

4. Real-time Analytics and IoT Applications

5. Geospatial Data

6. Horizontal Data Partitioning

Q2. State and Explain the features of MongoDB.

Ans.
 Here are some key features of MongoDB:

1. Document-Oriented:
MongoDB is a document-oriented database, where data is stored in the form of flexible and schema-less BSON documents (binary JSON). Each document represents a single record and can have varying fields, allowing for easy storage of complex and dynamic data.

2. Scalability:
MongoDB can scale horizontally, allowing you to distribute data across multiple servers or nodes. It supports sharding, a process of partitioning data across different machines to ensure high availability and handle large datasets.

3. High Performance:
MongoDB's document structure and indexing capabilities result in high read and write performance. It can efficiently handle real-time data, making it suitable for high-throughput applications.

4. Flexible Schema:
Unlike traditional relational databases, MongoDB has a flexible schema, allowing you to evolve the data model without affecting existing records. This feature is particularly useful during the development phase when requirements may change frequently.

5. Powerful Query Language:
MongoDB provides a rich query language with support for complex queries, sorting, filtering, geospatial queries, and aggregation operations. It allows you to retrieve and process data in various ways, facilitating advanced data analytics.
